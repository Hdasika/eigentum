# Downloading Images from each record of CSV link(Base url)

```python
import pandas as pd
import requests
import urllib.request
import os
from bs4 import BeautifulSoup

# Uncomment when running on colab
from google.colab import drive
drive.mount('/content/gdrive')
os.chdir('/content/gdrive/My Drive/733')
```

```python
# edited to code to avoid the overhead of downloading  and create the folders only when the folder does not exit
```

```python
realtor_properties_df = pd.read_csv('realtor_properties.csv')
total_properties = len(realtor_properties_df)

headers = {'Cookie': 'TermsOfUseAgreement=2018-06-07'}
base_img_folder = "images/"
pwd = os.getcwd()
i = 0

if not os.path.exists(base_img_folder):
    os.mkdir(base_img_folder)

for index, realtor_property in realtor_properties_df.iterrows():


    os.chdir(base_img_folder)
    image_folder_name = realtor_property['mls']
    j = 0

    if os.path.exists(image_folder_name):
        print('Folder exist', image_folder_name)
        i = i + 1
    else:
        print('Getting Images for {} record {} out of {}'.format(realtor_property['mls'], (index + 1), total_properties))

        realtorHtml = requests.get(realtor_property['absoluteUrl'], headers=headers)
        soup = BeautifulSoup(realtorHtml.text, 'html.parser')
        imageElements = soup.findAll('img', id=lambda x: x and x.startswith('propimg'))
        os.makedirs(image_folder_name)
        os.chdir(image_folder_name)
        
        for jpgLink in imageElements:
            j = j + 1
            file_name = realtor_property['mls'] + '_' + str(j) + '.jpg'
            # Saves the images in folder Path in MLS folder with mls.jpg
            urllib.request.urlretrieve(jpgLink['src'], file_name)
            # Uncomment when running on colab
            #!cp file_name /content/gdrive/My\ Drive/733/
    os.chdir(pwd)
```
