# Joining Multiple files - Interest

```python
import numpy as np
import pandas as pd
import os
bca_Intrest_dir = os.getcwd() + "/Interest/"
def combine_Interest_csv():
    Interest_Combine_df = pd.DataFrame(columns=['Property Type','Area','Price_Index','BenchMark_Price','BenchMark_Price_dol'])
    for file in os.listdir(bca_Intrest_dir):
        if file.endswith(".csv"):
            Month=file.split('_')[0]
            Year=file.split('_')[1].split('.')[0]
            Data = pd.read_csv(bca_Intrest_dir + file)
            Data[['Area','first_Half']]=Data['Area'].str.split('$',expand=True)
            Data['second_Half']=Data['Benchmark Price'].str.split(' ',expand=True)[0]
            Data['Price_Index']=Data['Benchmark Price'].str.split(' ',expand=True)[1]
            Data['BenchMark_Price_dol']='$'+Data['first_Half']+Data['second_Half']
            Data['BenchMark_Price']=Data['first_Half']+Data['second_Half']
            Data=Data[['Property Type','Area','Price_Index','BenchMark_Price','BenchMark_Price_dol']]
            Data['Year'],Data['Month']=Year,Month
            Interest_Combine_df=pd.concat([Interest_Combine_df,Data],ignore_index=True)
    return Interest_Combine_df
intrest=combine_Interest_csv()
intrest.to_csv('Interest_combined.csv')
```

# Joining BC Property Data file

```python
bca_property_tax_dir = os.getcwd() + "/bca_property_tax/"
bca_property_tax_combined_df = pd.DataFrame()

for file in os.listdir(bca_property_tax_dir):
    if file.endswith(".csv"):
        bc_assessment_df = pd.read_csv(bca_property_tax_dir + file)
        bc_assessment_df['PID']=bc_assessment_df['PID'].str.replace('-','')
        bc_assessment_df['HOUSE_PRICE'] =  bc_assessment_df.apply(lambda x: (x['CURRENT_LAND_VALUE']+x['CURRENT_IMPROVEMENT_VALUE']), axis = 1)
        bc_assessment_df = bc_assessment_df.dropna(subset=['PROPERTY_POSTAL_CODE', 'HOUSE_PRICE','PID'])
        bc_assessment_df = bc_assessment_df[~bc_assessment_df['PROPERTY_POSTAL_CODE'].str.startswith('M')]
        bc_assessment_df[['AREA_CODE', 'LOCALITY_CODE']] = bc_assessment_df['PROPERTY_POSTAL_CODE'] \
            .str.split(expand=True)
        #bc_assessment_df = bc_assessment_df.groupby('AREA_CODE')[['HOUSE_PRICE']].median().reset_index()
        bc_assessment_df['YEAR'] = file.title().split('.')[0]
        bc_assessment_df['PID']=bc_assessment_df['PID'].astype(int)
        bca_property_tax_combined_df = pd.concat([bca_property_tax_combined_df, bc_assessment_df],
                                                 ignore_index=True)
bca_property_tax_combined_df
```

# BOC Conversion

```python
boc_Interest = pd.read_csv('boc.csv')
boc_Interest['5y-bond'] = boc_Interest['5y-bond'].interpolate()
boc_Interest['Date']=boc_Interest['date'].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d').strftime('%B-%Y'))
boc_Interest_final=boc_Interest[['5y-bond','Date']]
boc_Interest_final=boc_Interest_final.dropna()
boc_Interest_fin=boc_Interest_final.groupby('Date')['5y-bond'].median().reset_index()
boc_Interest_fin=boc_Interest_fin.drop_duplicates()
```
